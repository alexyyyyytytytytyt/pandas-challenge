## Age Demographics

```python
group_names = ["<10", "10-14", "15-19", "20-24", "25-29", "30-34", "35-39", "40+"]
bins = [0,9,14,19,24,29,34,39,100]

purchase_data["Age Group"] = pd.cut(uniqueSN_data["Age"],bins, labels=group_names)
purchase_data

player_age = purchase_data["Age Group"].value_counts()
index_player = player_age.sort_index()
pd.DataFrame(index_player)
```

```python
summation = purchase_data["Age Group"].count()
summation
```

```python
ave = index_player/summation * 100
more_data = pd.DataFrame({"Total Count": index_player, "Percentage of Players": ave})
more_data["Percentage of Players"] = more_data["Percentage of Players"].map("{:.2f}%".format)
more_data
```
